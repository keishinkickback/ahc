```lua
torch.setdefaulttensortype('torch.FloatTensor')
agg = require 'aggclustering'
Plot = require 'itorch.Plot'

function plotClusters(x, groups, nClusters)
    plot = Plot()
    colortable ={'red', 'blue', 'green', 'yellow', 'black','purple'}
    for ci, g in ipairs(groups) do
        for _, p in ipairs(g) do
            colorIndex = ci % 6 + 1
            plot:circle(x[{{p},{1}}]:reshape(1), x[{{p},{2}}]:reshape(1), colortable[colorIndex])
            :text(x[{{p},{1}}]:reshape(1), x[{{p},{2}}]:reshape(1), tostring(p), torch.Tensor({0}), 'black')
        end
    end
    plot:title('Scatter Plot: ' ..tostring(nClusters) .. ' clusters')
    plot:xaxis('x_1'):yaxis('x_2')
    plot:draw()
end
```

```lua
x = torch.Tensor({ { 1,  1},
                   { 2,  1},
                   {-1, -1},
                   {-1, -2},
                   {-2, -2} })
x = x * 10
print('data')
print(x)
for i=5, 2, -1 do
    groups, dists, md = agg:clustering_nClusters(x, agg.Euclidean, i, 'single')
    print('\nnumber of clusters: '.. tostring(i))
    print('clusters')
    print(groups)
    print('cluster-wise distances')
    print(dists)
    plotClusters(x, groups, i)
end
```

```lua
x = torch.randn(15,2) * 10
groups, dists, md = agg:clustering_threshold(x, agg.Cosine, 0.1, 'centroid')
print(groups)
print(dists)
plotClusters(x, groups, #groups)
```
